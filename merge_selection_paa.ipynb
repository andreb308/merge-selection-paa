```python
from selectionSort import *
from mergeSort import *
import numpy as np
import time
import matplotlib
np.set_printoptions(threshold=np.inf, suppress=True)
```

```python
ARR_SIZES = np.array([50, 500, 5000, 50000])
NUMBER_OF_TRIES = 1 # Número de vezes que o algoritmo de ordenação será rodado (para calcular a média antes de plotar o gráfico)
mergeData = {
                '50-ascending': np.array([]),'50-descending': np.array([]),'50-random': np.array([]),'500-ascending': np.array([]),'500-descending': np.array([]),'500-random': np.array([]),'5000-ascending': np.array([]),'5000-descending': np.array([]),'5000-random': np.array([]),'50000-ascending': np.array([]),'50000-descending': np.array([]),'50000-random': np.array([])
}

selectionData = {
                '50-ascending': np.array([]),'50-descending': np.array([]),'50-random': np.array([]),'500-ascending': np.array([]),'500-descending': np.array([]),'500-random': np.array([]),'5000-ascending': np.array([]),'5000-descending': np.array([]),'5000-random': np.array([]),'50000-ascending': np.array([]),'50000-descending': np.array([]),'50000-random': np.array([])}
# key = f'{ARR_SIZE}-selection'
```

# Merge Sort

## Random Integers

```python
for size in ARR_SIZES:
    arr_test = []
    for i in range(NUMBER_OF_TRIES):
        random_array = np.random.randint(1, 100000001, size=size)

        start_time = time.time()
        mergeSortedList = mergeSort(random_array)
        elapsed_time = time.time() - start_time
    
        if elapsed_time:
            arr_test.append(elapsed_time)
 
        key = f'{size}-random'
        mergeData.update({key: np.mean(np.array([arr_test]))})
```

## Array of Ascending Integers

```python
for size in ARR_SIZES:
    arr_test = []
    
    # Generates the new array when the size changes
    array_f2b = np.array([i for i in range(size)])

    for i in range(NUMBER_OF_TRIES):

        start_time = time.time()
        mergeSortedList = mergeSort(array_f2b)
        elapsed_time = time.time() - start_time
    
        if elapsed_time:
            arr_test.append(elapsed_time)
 
        key = f'{size}-ascending'
        mergeData.update({key: np.mean(np.array([arr_test]))})
```

## Array of Descending Integers

```python
for size in ARR_SIZES:
    arr_test = []
    
    # Generates the new array when the size changes
    array_b2f = np.arange(size, 0, -1)

    for i in range(NUMBER_OF_TRIES):

        start_time = time.time()
        mergeSortedList = mergeSort(array_b2f)
        elapsed_time = time.time() - start_time
    
        if elapsed_time:
            arr_test.append(elapsed_time)
 
        key = f'{size}-descending'
        mergeData.update({key : np.mean(np.array([arr_test]))})
    
```

```python
import matplotlib.pyplot as plt



x_values = [50, 500, 5000, 50000]
ascending_values = []
descending_values = []
random_values = []

for key in mergeData:
    if 'ascending' in key:
        ascending_values.append(mergeData[key])
    elif 'descending' in key:
        descending_values.append(mergeData[key])
    elif 'random' in key:
        random_values.append(mergeData[key])

plt.plot(x_values, ascending_values, label='Ascending')
plt.plot(x_values, descending_values, label='Descending')
plt.plot(x_values, random_values, label='Random')

plt.xlabel('Número de Elementos')
plt.ylabel('Tempo (s)')
plt.title('Merge Sort: Nº de Elementos por Tempo')

plt.legend()
plt.show()
```
